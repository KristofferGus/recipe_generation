```python
import torch
import matplotlib.pyplot as plt
```

```python
def show_model(model_dict):

    plt.plot(model_dict["avg_gen_losses"])
    plt.xlabel("Batch nr")
    plt.ylabel("Loss")
    plt.show()
    print("Nr of parameters:", sum([t.numel() for n,t in list(model_dict["model"].items())]))
    print("Dropp rate",model_dict['drop_rate'])
    print("Learning rate",model_dict['gen_lr'])
    print("batch size",model_dict['batch_size'])
    print("sequance length", model_dict['seq_len'])
    print("Number of batches trained", model_dict['nr_batches'])





```

```python
#The normal model
#There are two perplexeties, first is of both ingridents and directions, second is only for directions.
#Observ that the model is much better att perdicting ingridents than directions.
good_simpel_model = torch.load("simpel_model2.pth")
show_model(good_simpel_model)
perplexeties = torch.load("simpel_model2_perplexities_rec.pth")
perplexeties2 = torch.load("simpel_model2_perplexities.pth")
print("Perplexity for directions:", sum(perplexeties) / len(perplexeties))
print("Perplexity for all:", sum(perplexeties2) / len(perplexeties2))
```
